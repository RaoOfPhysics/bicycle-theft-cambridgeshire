```r
library(ggmap)
library(dplyr)
```

```r
df <- read.csv("../data/processed/bikethefts.csv")
```

Longitude and latitude of Cambridge

```r
myLocation <- c(lon = 0.1218, lat = 52.2053)
```

get map using get_map and plot using ggmap(). Level of zoom can be set with zoom, with higher numbers being more zoomed. 

```r
get_map(location=myLocation, zoom = 15) %>% ggmap() +
  geom_point(data = df, aes(x = Longitude, y = Latitude), color = 'red', size = 1)
```

```r
get_map(location=myLocation, zoom = 13) %>% ggmap() +
  geom_point(data = df, aes(x = Longitude, y = Latitude), color = 'red', size = 1)
```
